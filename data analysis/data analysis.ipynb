#如何将多个工作表的数据写入到一个excel中
import pandas as pd
filename='../data/happiness.xlsx'
df_data=pd.read_excel(filename,sheet_name=['2015','2017'])
top5_2015=df_data['2015'].head()
top5_2017=df_data['2017'].head()
writer=pd.ExcelWriter('../../tmp/2015_2017_top5.xlsx')
top5_2015.to_excel(writer,'2015 top 5')
top5_2017.to_excel(writer,'2017 top 5')
writer.save()
